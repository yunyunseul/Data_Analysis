

1. 메뉴 데이터

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/A. 개인 프로젝트/dvthedv_data - 메뉴.csv')
df1 = df[['메뉴명', '해금조건', '관련파티', '판매가격_Lv1', '재료명_1', '재료량_1', '맛점수최소', '맛점수최대', '생산량최소', '생산량최대']]
df2 = df[['메뉴명', '해금조건', '관련파티', '판매가격_Lv1', '재료명_2', '재료량_2', '맛점수최소', '맛점수최대', '생산량최소', '생산량최대']]
df3 = df[['메뉴명', '해금조건', '관련파티', '판매가격_Lv1', '재료명_3', '재료량_3', '맛점수최소', '맛점수최대', '생산량최소', '생산량최대']]
df4 = df[['메뉴명', '해금조건', '관련파티', '판매가격_Lv1', '재료명_4', '재료량_4', '맛점수최소', '맛점수최대', '생산량최소', '생산량최대']]
df1 = df1.rename(columns={'재료명_1' : '재료명', '재료량_1': '재료량'})
df2 = df2.rename(columns={'재료명_2' : '재료명', '재료량_2': '재료량'})
df3 = df3.rename(columns={'재료명_3' : '재료명', '재료량_3': '재료량'})
df4 = df4.rename(columns={'재료명_4' : '재료명', '재료량_4': '재료량'})

menu = pd.concat([df1, df2, df3, df4])
menu = menu.dropna(subset = ['재료명', '재료량'])
menu = menu.sort_values(by = '메뉴명')

menu['재료명'] = menu['재료명'].str.strip()
menu['메뉴명'] = menu['메뉴명'].str.strip()

print(menu.head(10))
```

```python
# 강화 레벨별 판매 가격 = 기본가격 + 기본가격 * 0.3 * (레벨 - 1), 소수점 버림
# 트러플, 히알로네마 메뉴는 레벨 5까지 강화, 나머지 메뉴는 레벨 10까지 강화 가능
# 트러플은 장고로부터 개당 650 주고 구매하는게 유일한 입수 경로. 수익에서 원가 가격을 빼서 계산

def max_price(x):
  if x['재료명'] == '트러플':
     a = (x['판매가격_Lv1'] + x['판매가격_Lv1'] * 0.3 * 4) - (x['재료량'] * 650)
  elif x['재료명'] == '히알로네마':
     a = x['판매가격_Lv1'] + x['판매가격_Lv1'] * 0.3 * 4
  else:
     a =  x['판매가격_Lv1'] + x['판매가격_Lv1'] * 0.3 * 9
  return a

menu['판매가격_Lvmax'] = menu.apply(max_price, axis = 1).astype('int')


#트러플, 히알로네마 요리 예시
print(menu.loc[menu['메뉴명'] == '트러플 블루랍스터 꼬리회'])
print(menu.loc[menu['메뉴명'] == '히알로네마 아귀찜'])
print(menu.loc[menu['메뉴명'] == '감태 초무침'])
```

```python
diffi = pd.read_csv('/content/drive/MyDrive/A. 개인 프로젝트/dvthedv_data - 재료 난이도.csv')
lc = pd.read_csv('/content/drive/MyDrive/A. 개인 프로젝트/dvthedv_data - 재료 위치.csv')
```

```python
lc.loc[(lc['양식불가여부'] == '양식불가') & (lc['대분류'] == '생물'), '대분류'] = '양식불가어종'
lc.loc[(lc['양식불가여부'] != '양식불가') & (lc['대분류'] == '생물'), '대분류'] = '양식가능어종'
```

4. 지표 생성

```python
all['필요노력'] = all['재료량'] * all['난이도']
```

```python
effortMenu = all.groupby('메뉴명').agg(
    {'필요노력' : 'sum',
     '생산량최대' : 'min',
     '생산량최소' : 'min',
     '판매가격_Lvmax' : 'min'}
).reset_index()

effortMenu.head()

#print(effortMenu.loc[menu['메뉴명'] == '트러플 블루랍스터 꼬리회'])
print(effortMenu.loc[effortMenu['메뉴명'] == '히알로네마 아귀찜'])
```

```python
effortMenu['필요노력'] = effortMenu['필요노력'].astype('float')
effortMenu.loc[effortMenu['생산량최대'] == '?', '생산량최대'] = effortMenu.loc[effortMenu['생산량최대'] == '?', '생산량최소']
# 생산량최대값이 없는 '핫페퍼튜나' 메뉴는 생산량최소값으로 대체

effortMenu['생산량최대'] = effortMenu['생산량최대'].astype('float')

effortMenu['접시당필요노력'] = effortMenu['필요노력'] / effortMenu['생산량최대']
effortMenu['접시당가격'] = effortMenu['판매가격_Lvmax'] / effortMenu['생산량최대']
effortMenu['필요노력당가격'] = effortMenu['접시당가격'] / effortMenu['접시당필요노력']
effortMenu['가격당필요노력'] = effortMenu['접시당필요노력'] / effortMenu['접시당가격']

effortMenu.head()
```

5. 최적의 메뉴 찾기
- 노력의 최소화
- 수익의 최대화

```python
# 필요노력이 적은 메뉴
effortMenu.sort_values(by = '접시당필요노력', ascending = True).head(10)   # 한 접시를 만들 때 필요한 노력 정도
```

```python
effortMenu.sort_values(by = '가격당필요노력', ascending = True).head(10)  # 한 접시를 만들 때 1원을 받기 위해 필요한 노력 정도
```

```python
effortMenu['일최대수익'] = effortMenu['판매가격_Lvmax'] * 45 / effortMenu['생산량최대']   #플레티넘 기준. 야간 다이빙 하지 않는다고 가정
# 최대 손님 수 - 골드: 28명, 야간 다이빙 시 21명 / 플래티넘: 36명, 야간다이빙 시 22명 / 다이아: 45명, 야간다이빙 시 27

effortMenu['일최대수익을 위한 메뉴수']  = 45 / effortMenu['생산량최대']
effortMenu['일최대수익을 위한 노력'] = effortMenu['일최대수익을 위한 메뉴수'] * effortMenu['필요노력']
effortMenu.sort_values(by = '일최대수익을 위한 노력', ascending = True).head(10)  # 해당 메뉴로 통일할 때 필요한 노력
```

```python
# 수익이 높은 메뉴
effortMenu.sort_values(by = '필요노력당가격', ascending = False).head(10) # 한 단위 노력을 했을 때 얻을 수 있는 수익
```
